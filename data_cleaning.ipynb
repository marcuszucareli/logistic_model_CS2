## Maps

```python
# Add column saying wich side won the round (0 for CT and 1 for TR)
ct_wins_types = ['ct_win.svg', 'bomb_defused.svg', 'stopwatch.svg']
df['win_side'] = df['win_type'].apply(lambda x: 0 if x in ct_wins_types else 1)

df = df.drop('winner', axis=1)

# Rounds per map
maps_stats = df.groupby('map_name').agg(
    total_rounds_played = ('match_url', 'count'),
    total_matches_played = ('match_url', 'nunique'),
    tr_wins=('win_side', 'sum')
)
maps_stats['avg_rounds_per_map'] = maps_stats['total_rounds_played'] / maps_stats['total_matches_played']
maps_stats['ct_wins'] = maps_stats['total_rounds_played'] - maps_stats['tr_wins']
maps_stats['ct_tr_ratio'] = maps_stats['ct_wins'] / maps_stats['tr_wins']

maps_stats
```

# Feature Engineering

In this section we will create variables derivated from our data set that might
help a good fit for the model, they are:

- equipament_ct and equipament_tr: equipament value for sides, instead of teams
- equipament_ratio: equipament_ct / equipament_tr
- equipament_diff: equipament_ct - equipament_tr
- is_pistol: if the round is a pistol round

```python
# Create equipaments_ct and equipaments_tr 
def get_side_equipament(team_1_side, equipament_1, equipament_2):
    
    equipament_ct = equipament_1 if team_1_side == 0 else equipament_2
    equipament_tr = equipament_1 if team_1_side == 1 else equipament_2
    
    return equipament_ct, equipament_tr

df[['equipament_ct', 'equipament_tr']] = df.apply(
    lambda row: get_side_equipament(
        row['team_1_side'],
        row['equipament_1'],
        row['equipament_2']
    ),
    axis=1,
    result_type='expand'
)

df = df.drop(
    [
        'equipament_1',
        'equipament_2',
        'team_1',
        'team_2'
    ],
    axis=1
)

df
```

```python
# Equipament variables
df['equipament_ratio'] = df['equipament_ct'] / df['equipament_tr']
df['equipament_dif'] = df['equipament_ct'] - df['equipament_tr']

df
```

```python
# is_pistol
'''
A pistol round occurs everytime a teams change sides (except if it's overtime),
or the map changes.
Overtime begins when the game reachs 12 x 12 => 24 rounds
'''
is_same_map = (df['map_name'] == df['map_name'].shift(1))
is_same_game = (df['match_url'] == df['match_url'].shift(1))
is_different_side = (df['team_1_side'] != df['team_1_side'].shift(1))
is_overtime = (df['round'] > 24)

df['is_pistol'] = (
    ~is_overtime & is_different_side | \
    ~is_same_map | \
    ~is_same_game
).astype(int)

df
```
